```python
#import packages
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#import csv and only keep important columns
df = pd.read_csv("data/220511_monatszahlenmonatszahlen2204_verkehrsunfaelle.csv")
df = df.iloc[:,:5]
df
```

```python
df.info()
```

```python
#check for null values and drop null-value rows
df[df["WERT"].isnull()].head()
```

```python
df = df.dropna()
df.info()
#df.describe()
```

```python
#check unique types of categories and accident types
categories=list(set(list(df["MONATSZAHL"])))
accidents=list(set(list(df["AUSPRAEGUNG"])))
categories
```

```python
accidents
```

```python
#divide the dataset into two dataframes
df_summe = df[df['MONAT'] == 'Summe'] # dataframe with summe only
df_no_summe = df[df['MONAT'] != 'Summe'] # df without summe
df = df_no_summe

#change the date format in MONAT column
df['MONAT'] = pd.to_datetime(df['MONAT'],format='%Y%m')
df
```

```python
df_summe
```

```python
#only keep the insgesamt values in both dataframes, remove other accident types
df_summe_insgesamt = df_summe[df_summe['AUSPRAEGUNG'] == 'insgesamt']
df = df[df['AUSPRAEGUNG'] == 'insgesamt']
```

```python
#plot of total accidents across all categories per year
fig, ax = plt.subplots(figsize=(19,6))
df_summe_insgesamt.groupby(['JAHR', 'MONATSZAHL']).sum()['WERT'].unstack().plot(kind='bar', colormap = 'tab20', ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Accidents')
fig.suptitle('Total Number of Accidents Across All Categories (Per Year)')
fig.savefig('plot_bar.jpg')
plt.show()
```

```python
#stacked plot of total accidents across all categories per year
fig, ax = plt.subplots(figsize=(19,6))
df_summe_insgesamt.groupby(['JAHR', 'MONATSZAHL']).sum()['WERT'].unstack().plot(kind='bar', colormap = 'tab20', stacked=True, ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Accidents')
fig.suptitle('Total Number of Accidents Across All Categories (Per Year)')
fig.savefig('plot_bar_stacked_v.jpg')
plt.show()
```

```python
#line plot of total accidents across all categories per year
fig, ax = plt.subplots(figsize=(16,6))
df_summe_insgesamt.groupby(['JAHR', 'MONATSZAHL']).sum()['WERT'].unstack().plot(kind='line', colormap = 'tab20', ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Accidents')
fig.suptitle('Total Number of Accidents Across All Categories (Per Year)')
fig.savefig('plot_line.jpg')
plt.show()
```

```python
#plot of total accidents across all categories per month
fig, ax = plt.subplots(figsize=(16,6))
df.groupby(['MONAT', 'MONATSZAHL']).sum()['WERT'].unstack().plot(kind='line', colormap = 'tab20', ax=ax)
ax.set_xlabel('Month')
ax.set_ylabel('Number of Accidents')
fig.suptitle('Total Number of Accidents Across All Categories (Per Month)')
fig.savefig('plot_line_month.jpg')
plt.show()
```

```python
#plot of total accidents across all categories per month (last 4 years)
fig, ax = plt.subplots(figsize=(16,6))
df.groupby(['MONAT', 'MONATSZAHL']).sum()['WERT'].unstack().plot(kind='line', colormap = 'tab20', ax=ax)
ax.set_xlabel('Month')
ax.set_ylabel('Number of Accidents')
ax.set_xlim(pd.Timestamp('2018'),pd.Timestamp('2022'))
fig.suptitle('Total Number of Accidents Across All Categories in Last 4-Years')
fig.savefig('plot_line_month_last4yrs.jpg')
plt.show()
```
