```python
df = pd.read_csv("data/220511_monatszahlenmonatszahlen2204_verkehrsunfaelle.csv")
df = df.iloc[:,:5]
df
```

```python
df.info()
```

```python
df[df["WERT"].isnull()]
```

```python
df = df.dropna()
df.info()
#df.describe()
```

```python
categories=list(set(list(df["MONATSZAHL"])))
accidents=list(set(list(df["AUSPRAEGUNG"])))
categories
```

```python
accidents
```

```python
df_summe = df[df['MONAT'] == 'Summe'] # dataframe with summe only
df_no_summe = df[df['MONAT'] != 'Summe'] # df without summe
df = df_no_summe
df
```

```python
df_summe
```

```python
df_summe_insgesamt = df_summe[df_summe['AUSPRAEGUNG'] == 'insgesamt']
df_summe_insgesamt
```

```python
fig, ax = plt.subplots(figsize=(16,6))
df_summe_insgesamt.groupby(['JAHR', 'MONATSZAHL']).mean()['WERT'].unstack().plot(ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Number of accidents')
fig.suptitle('Yearly number of accidents per category')
fig.savefig('plot_line.jpg')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(19,6))
df_summe_insgesamt.groupby(['JAHR', 'MONATSZAHL']).mean()['WERT'].unstack().plot(kind='bar', ax=ax)
#df_groups.plot(kind='bar')

ax.set_xlabel('Year')
ax.set_ylabel('Number of accidents')
fig.suptitle('Yearly number of accidents per category')
fig.savefig('plot_bar.jpg')
plt.show()
```

```python
df_grouped = df_summe_insgesamt.groupby(['JAHR', 'MONATSZAHL']).mean()['WERT'].unstack()
fig, ax = plt.subplots(figsize=(19,6))

left = len(df_grouped) * [0]
for idx, name in enumerate(categories):
    plt.barh(df_grouped.index, df_grouped[name], left = left)
    left = left + df_grouped[name]

ax.set_xlabel('Total Number of accidents')
ax.set_ylabel('Year')
plt.legend(categories, bbox_to_anchor=([0.55, 1, 0, 0]), ncol=4, frameon=False)
fig.suptitle('Yearly number of accidents across all categories')
fig.savefig('plot_bar_stacked_h.jpg')
plt.show()
```

```python
#df.assign(dummy = 1).groupby(
#  ['dummy','state']
#).size().to_frame().unstack().plot(kind='bar',stacked=True,legend=False)


fig, ax = plt.subplots(figsize=(19,6))
df_summe_insgesamt.groupby(['JAHR', 'MONATSZAHL']).mean()['WERT'].unstack().plot(kind='bar', stacked=True, ax=ax)
#df_groups.plot(kind='bar')

ax.set_xlabel('Year')
ax.set_ylabel('Number of accidents')
fig.suptitle('Total Number of accidents')
fig.savefig('plot_bar_stacked_v.jpg')
plt.show()
```
